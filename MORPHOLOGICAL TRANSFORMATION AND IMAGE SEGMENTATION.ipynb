```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

# DILATION

```python
img = cv.imread("C:/Users/SN112/Downloads/black.jpg",0)
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
_,mask= cv.threshold(img,220,225,cv.THRESH_BINARY)
kernel = np.ones((3,3),np.uint8)
dilation = cv.dilate(mask,kernel,iterations=1)
erode = cv.erode(mask,kernel,iterations=1)
titles = ['image','mask','dilate','erode']
images = [img,mask,dilation,erode]
for i in range(len(images)):
        plt.subplot(2,2,i+1),plt.imshow(images[1])
        plt.title(titles[1])
        plt.xticks([]),plt.yticks([])
```

# MORPHOLOGICAL OPENING =ERROSION FOLLOWED BY DILATION


```python
open = cv.morphologyEx(img,cv.MORPH_OPEN,kernel)
```

```python
cv.imshow('img',img)
cv.imshow('morph open',open)
cv.waitKey(0)
cv.destroyALLWindows()
```

# MORPHOLOGICAL GRADIENT- ERROSION,DILATION,DIFFERENCE OF BOTH

```python
gradient = cv.morphologyEx(img,cv.MORPH_GRADIENT,kernel)
```

# DISPLAY - MORPHOLOGICAL GRADIENT

```python
cv.imshow('ORIGINAL',img)
cv.imshow('MORPH GRADIENT',gradient)
cv.waitKey(0)
cv.destroyAllWindows()
```

# TOPHAT - OPENING AND DIFFERNCE OF ORGINAL AND OPENING

```python
tophat = cv.morphologyEx(img,cv.MORPH_TOPHAT,kernel)
```

```python
cv.imshow('ORIGINAL',img)
cv.imshow('TOP HAT',tophat)
cv.waitKey(0)
cv.destroyAllWindows()
```

# BLACK HAT

```python
blackhat = cv.morphologyEx(img,cv.MORPH_BLACKHAT,kernel)
```

```python
cv.imshow('ORIGINAL',img)
cv.imshow('BLACK HAT',blackhat)
cv.waitKey(0)
cv.destroyAllWindows()
```
